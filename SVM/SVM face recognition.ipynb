```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_lfw_people
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
```

```python
faces.target_names
```

```python
faces.images.shape
```

```python
fig, ax = plt.subplots(5,5, figsize=(16,9))
for i, ax_i in enumerate(ax.flat):
    ax_i.imshow(faces.images[i], cmap="bone")
    ax_i.set(xticks=[],yticks=[],xlabel=faces.target_names[faces.target[i]])
```

```python
from sklearn.svm import SVC
from sklearn.decomposition import RandomizedPCA
from sklearn.pipeline import make_pipeline
```

```python
pca = RandomizedPCA(n_components=150, whiten=True,random_state=42)
svc = SVC(kernel="rbf",class_weight="balanced")
model = make_pipeline(pca,svc)
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(faces.data, faces.target, random_state=42)
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
param_grid = {
    "svc__C":[0.1,1,4,5,6,9,1000],
    "svc__gamma":[0.0001,0.0002,0.0004,0.01]
}
grid = GridSearchCV(model,param_grid)
%time grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
classifier = grid.best_estimator_
y_fit = classifier.predict(X_test)
```

```python
fig, ax = plt.subplots(5,9,figsize=(16,9))

for i, ax_i in enumerate(ax.flat):
    ax_i.imshow(X_test[i].reshape(62,47), cmap="bone")
    ax_i.set(xticks=[], yticks=[])
    ax_i.set_ylabel(faces.target_names[y_fit[i]].split()[-1],
                   color="black" if y_fit[i]==Y_test[i] else "red")
fig.suptitle("Incorrect in red")
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,y_fit,target_names=faces.target_names))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
mat = confusion_matrix(Y_test,y_fit)
print(mat)
```

```python
import seaborn as sns; sns.set()
```
