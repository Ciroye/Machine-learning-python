```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## By category

```python
gender = ["Male", "Female"]
social_level = ["Poor","Middle Class", "Rich"]
```

```python
n = 500 #Data amount
gender_data = []
social_level_data = []
for i in range(n):
    gender_data.append(np.random.choice(gender))
    social_level_data.append(np.random.choice(social_level))
```

```python
gender_data = pd.DataFrame(gender_data)
gender_data.head(5)
```

```python
social_level = pd.DataFrame(social_level_data)
social_level.tail(2)
```

```python
height = 160 + 30 * np.random.randn(n)
weight = 75 + 25 * np.random.randn(n)
age = 30 + 14 * np.random.randn(n)
income = 18000 + 4000 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "height":height,
        "weight":weight,
        "age":age,
        "income":income,    
    }
)
```

```python
data["Gender"] = gender_data
data["Social level"] = social_level
```

```python
data.head(5)
```

```python
group_by_gender = data.groupby("Gender")
```

```python
group_by_gender.groups
```

```python
for name, groups in group_by_gender:
    print groups
```

```python
double_group = data.groupby(["Gender","Social level"])
double_group.groups
```

```python
for names,groups in double_group:
    print names
    print groups
```

## Filter data

```python
double_group["age"].filter(lambda x: x.sum > 2400)
```

## Transformation variables

```python
zscore = lambda x: (x-x.mean())/x.std()
```

```python
zgroup = double_group.transform(zscore)
```

```python
%matplotlib inline
```

```python
plt.hist(zgroup["age"])
```

## Fill missing values

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```
