```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Uniform distribution

```python
a = 1
b = 2
n = 10000000
data = np.random.uniform(a,b,n)
```

```python
%matplotlib inline
```

```python
plt.hist(data)
```

# Normal distribution

```python
data = np.random.randn(100)
```

```python
x = np.arange(100)
plt.plot(x, data)
```

```python
%matplotlib inline
plt.hist(data)
```

```python
me = 5
sd = 2
data = me + (sd*np.random.randn(1999)) # Z = X - m / sd ->N(0,1) , X = m + sd*X
plt.hist(data)
```
