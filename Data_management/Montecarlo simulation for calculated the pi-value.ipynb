```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

* Generate two random numbers between 0 and 1 n times
* Calculated x*x + y*y
    * if value < 1:
        * in circle
    * else:
        * without circle
* Calculated numbers in circle divided by all the area
* Used this probability by calculated π (n experiment)
* Calculated the average for π

```python
def pi_montecarlo(n,experiment):
    pi_average = 0
    pi_value_list = []
    for i in range(experiment):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = (x[j]*x[j] + y[j]*y[j])
            if z < 1:
                value += 1
        float_value= float(value)
        pi_value = float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_average += pi_value
    pi = pi_average / experiment
    plt.plot(pi_value_list)
    return pi
```

```python
pi_montecarlo(1000000,200)
```
