```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../../datasets/ecom-expense/Ecom Expense.csv")
```

```python
df.head()
```

```python
dummy_gender = pd.get_dummies(df["Gender"],prefix = "Gender")
dummy_tier = pd.get_dummies(df["City Tier"],prefix = "City")
```

```python
dummy_gender.head()
```

```python
columns_name = list(df.columns.values)
columns_name
```

```python
df_new = df[columns_name].join(dummy_gender)
columns_name = df_new.columns.values.tolist()
df_new = df_new[columns_name].join(dummy_tier)
columns_name = df_new.columns.values.tolist()
df_new
```

```python
feature_cols = ["Monthly Income","Transaction Time",
               "Gender_Female","Gender_Male",
               "City_Tier 1","City_Tier 2","City_Tier 3",
                "Record"
               ]
```

```python
X = df_new[feature_cols]
Y = df_new["Total Spend"]
lm = LinearRegression()
lm.fit(X,Y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
list(zip(feature_cols,lm.coef_))
```

```python
lm.score(X,Y)
```

## Delete redundant dummy variable

```python
dummy_gender = pd.get_dummies(df["Gender"],prefix="Gender").iloc[:,1:]
```

```python
dummy_gender.head()
```

```python
dummy_tier = pd.get_dummies(df["City Tier"],prefix="City").iloc[:,1:]
dummy_tier.head()
```

```python
columns_name = list(df.columns.values)
df_new = df[columns_name].join(dummy_gender).join(dummy_tier)
df_new.head()
```

```python
feature_cols = ["Transaction Time","Monthly Income","Record","Gender_Male","City_Tier 2","City_Tier 3"]
X = df_new[feature_cols]
Y = df_new["Total Spend"]
```

```python
lm = LinearRegression()
lm.fit(X,Y)
```

```python
lm.intercept_
```

```python
list(zip(feature_cols,lm.coef_))
```

```python
lm.score(X,Y)
```
