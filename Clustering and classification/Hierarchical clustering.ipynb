```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../datasets/movies/movies.csv",sep=";")
```

```python
df
```

```python
Z = linkage(df,"single")
plt.figure(figsize=(25,10))
plt.title("Hierarchical dendrogram")
plt.xlabel("ID users")
plt.ylabel("Distance")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=25)
plt.show()
```

```python
Z = linkage(df,"ward")
plt.figure(figsize=(25,10))
plt.title("Hierarchical dendrogram")
plt.xlabel("ID users")
plt.ylabel("Distance")
dendrogram(Z,leaf_rotation=90.,leaf_font_size=25)
plt.show()
```
